```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# 데이터 불러오기
df = pd.read_csv('skin_irritation_cleaned.csv')
df.shape

# 컬럼 확인
df.columns

# SMILES 컬럼 확인
df['SMILES']

# 첫 번째 SMILES 확인
smiles_example = df['SMILES'][0]
smiles_example

# SMILES → Mol 변환
mol_example = Chem.MolFromSmiles(smiles_example)
mol_example

# 타입 확인
type(mol_example)

# Descriptor 계산
descriptor_example = Descriptors.CalcMolDescriptors(mol_example)
descriptor_example

# 결합이 끊어진 SMILES 확인 (예시)
df['SMILES'][3]
Chem.MolFromSmiles(df['SMILES'][3])

# Descriptor 저장 리스트
descriptor_list = []

# df의 각 행을 한줄씩 꺼내서 반복해보자
# 즉 첫째 줄, 둘째 줄, 셋째 줄 이렇게 하나씩 검사해보자
for index, row in df.iterrows():
    smiles = row['SMILES']  
# SMILES 칼럼의 행 값만 가져와서 변수 smiles에 저장하기

    # Quiz 3: 단일 구조만 남기기
 
    if '.' in smiles:
        continue  # 여러 분자면 건너뛰기

    # SMILES → Mol 변환
    mol = Chem.MolFromSmiles(smiles)

    # Mol 변환 실패 시 제외해야하니 건너뛰기
    if mol is None:
        continue

    # Descriptor 계산
    descriptor = Descriptors.CalcMolDescriptors(mol)

    # 계산한 descriptor을 리스트에 저장. 이 직전 코드에선 빈 리스트였음!
    descriptor_list.append(descriptor)

# 첫번째 분자 검문 -> 건너뛰기 없으면 mol로 바꿔 -> descriptor 저장 -> 두번째 분자 검문

# 만든 리스트를 보기 편하게 데이터프레임으로 만들자.
descriptor_df = pd.DataFrame(descriptor_list)

# 이제 최종적으로 남으 데이터 개수 출력하자
print("단일 구조만 남은 데이터 개수:", len(descriptor_df))
```
